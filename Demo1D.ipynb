# Adaptive Sampling to Reduce Epistemic Uncertainty Using Prediction Interval-Generation Neural Networks

We created a synthetic dataset with varying PI widths that consists of a sinusoid with Gaussian noise.
Specifically, the dataset contains 200 points which were generated using the equation 
$y = 5 \, \text{cos}(x) + 10 + \epsilon$, where $x \in [-5, 5]$ and $\epsilon$ is a 
Gaussian noise whose amplitude depends on $x$: $\epsilon = (2 \, \text{cos}(1.2 \, x) + 2) \, g$ and $g \sim \mathcal{N}(0, 1)$. 
Consider $x$ is a discrete variable that can take 100 values.
